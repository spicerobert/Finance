```python
import requests
from bs4 import BeautifulSoup as bf
url="https://web.pcc.gov.tw/prkms/tender/common/bulletion/readBulletion?querySentence=%E8%81%AF%E8%AA%BC%E6%B4%BB%E5%8B%95&tenderStatusType=%E6%8B%9B%E6%A8%99&sortCol=TENDER_NOTICE_DATE&timeRange=114&pageSize=10"
data=requests.get(url)
data=bf(data.text,'html.parser')
data
```

```python
data.select('table')[4].select('tbody')[0].select('tr')[9].select('td')
```

```python
n=0
for d in data.select('table'):
    key=d.text.find("新北市新店區公所")
    if key != -1:
        print(n)
    n+=1
```

```python
data.find_all('tr',class_=['tb_b2','tb_b3'])
```

```python
import pandas as pd
dict_comp={"單位名稱":[],"案號":[]}
for x in range(len(data.select('table')[4].select('tbody')[0].select('tr'))):
    company=data.select('table')[4].select('tbody')[0].select('tr')[x].select('td')[2]
    num=data.select('table')[4].select('tbody')[0].select('tr')[x].select('td')[3]
    dict_comp["單位名稱"].append(company.get_text(strip=True))
    dict_comp["案號"].append(num.get_text(strip=True))
d=pd.DataFrame(dict_comp)
d.to_csv("標案紀錄.csv",encoding='utf-8-sig')
```
